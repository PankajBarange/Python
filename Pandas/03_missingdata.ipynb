## MISSING DATA 

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1,2,3,4,5,6],
    'B': [1,2,3,4,5,6],
    'C': [1,2,np.nan,4,5,6],
    'D': [1,np.nan,np.nan,4,5,6]
}
    
```

```python
df = pd.DataFrame(data)
df 
```

```python
df.isna()
```

```python
df.isna().count()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
df.isna().any(axis = 1)
```

*  REMOVING MISSING VALUES

```python
df
```

```python
df.dropna()  # Drops all rows that have any NaN values
```

```python
df.dropna(thresh=2)   # Drops all rows that have less than 2 non-NaN values
```

* FILLING THE MISSING VALUES

```python
df
```

```python
df.fillna(0) # Fills all NaN values with 0
```

If I want to apply these I Have to write Implace=True



```python
df
```

```python
values = {
    'A': 0,
    'B': 100,
    'C': 200,
    'D': 300
}
df.fillna(value=values) # Fills NaN values with different values for each column
```
