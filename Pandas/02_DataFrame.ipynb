## DATAFRAME

- Combinations of Series

```python
import pandas as pd
import numpy as nb
```

```python
data = {
    "Name":[ "Pankaj", "Rahul","Aditya"],
    "Age": [22,21,23],
    "Town": ["Nagpur","Mumbai","Pune"],
    "Salary": [25000,30000,28000 ]
}
df =pd.DataFrame(data)
df
```

```python
data_list = [
    ["Pankaj", 22, "Nagpur", 25000],
    ["Rahul", 21, "Mumbai", 30000],
    ["Aditya", 23, "Pune", 28000 ]
]

Columns = ["Name","Age","Town","Salary"]
df2 = pd.DataFrame(data_list, columns=Columns)
df2
```
