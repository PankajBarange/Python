```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("anime.csv")
df.head()
```

```python
df.loc[1]["Title"]
```

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i ==")":
            break
        if i == "(":
            check = True
            continue
        if check:
            data += i
    return data
```

```python
# Make a New Column for Episode Count
df["Episodes"] = df["Title"].apply(extract_episodes)
df.head()
```

```python
df["Episodes"] = df["Episodes"].str.replace(" eps","")
df.head()
```

```python
df.loc[0]["Episodes"]
```

- Now Our Job Is To Convert Str into Int

```python
df["Episodes"] = df["Episodes"].astype(int)
df.head()
```

```python
# Make a New Column for Time Stamp
df.loc[0]["Title"]
```

```python
def extract_Timestamp(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] ==")":
            for j in range(i+1,i+20):
                data += txt[j]
            return data

```

```python
df["Timestamp"] = df["Title"].apply(extract_Timestamp)
df.head()
```

```python
# Which anime has the maximum number of episodes?
df[df["Episodes"] == df["Episodes"].max()]["Title"]
```

```python
# Which anime has the minimum number of episodes?
df[df["Episodes"] == df["Episodes"].min()]["Title"]
```

```python
#which anime has the highest score
df.loc[df["Score"].idxmax()]
```

```python
#give me top 5 highest scoring anime
df.sort_values(by="Score",ascending=False).head()
```

```python
#which anime has the highest episode count
df.loc[df["Episodes"].idxmax()]
```

```python
#animes with top 5 episode count
df.sort_values(by="Episodes",ascending=False).head()
```

```python
#which is the longest running anime
df.loc[df["Timestamp"].idxmax()]
```
